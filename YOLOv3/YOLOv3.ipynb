```python
# import relevant libraries
import cv2
import matplotlib.pyplot as plt
import utils
import darknet
import glob
import numpy as np

from utils import *
from darknet import Darknet
```

## Load required files 

```python
# set name cfg file
cfg_file = './drive/MyDrive/cfg/yolov3.cfg'

# set name of the pre-trained weights file
weight_file = './drive/MyDrive/weights/yolov3.weights'

# set name of coco names files
namesfile = './drive/MyDrive/data/coco.names'

# load the architecture
m = Darknet(cfg_file)

# load the weights
m.load_weights(weight_file)

# load classication classes
class_names = load_class_names(namesfile)
```

## View the Neural Network used in YOLOv3

```python
m.print_network()
```

## Data Processing

```python
human_imgs = []

# Load the image
img_paths = glob.glob('drive/My Drive/Data_images/test/*.jfif')

for path in img_paths: human_imgs.append(cv2.imread(path))

# Convert the image to RGB
human_imgs_rgb = [cv2.cvtColor(img, cv2.COLOR_BGR2RGB) for img in human_imgs]

# We resize the image to the input width and height of the first layer of the network.  
human_imgs_rs = [cv2.resize(img,(m.width, m.height)) for img in human_imgs]

# Set the IOU threshold. Default value is 0.4
iou_thresh = 0.4

# Set the NMS threshold. Default value is 0.6
nms_thresh = 0.6

```

## Testing network on all human test images

```python
# Detect objects in the image
for img in human_imgs_rs:
  boxes = detect_objects(m, img, iou_thresh, nms_thresh)

  # Print the objects found and the confidence level
  print_objects(boxes, class_names)

  #Plot the image with bounding boxes and corresponding object class labels
  plot_boxes(img, boxes, class_names, plot_labels = True)
```
